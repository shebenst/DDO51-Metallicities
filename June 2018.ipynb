>### Just as a test of the new cross-matched data set, I plot W1-W2 (WISE) vs. BP-DDO51 (APOGEE/GAIA) below.

```python
#Import libraries and functions
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
from scipy import stats
from mpl_toolkits.axes_grid1 import make_axes_locatable
from astropy.io import fits
from matplotlib import colors
%matplotlib nbagg

#Define variable and calculate W1-W2
w1_w2 = cat['W1mag'] - cat['W2mag']

#Define variables for BP-DDO51
bp_dd = cat['BPD']

#Construct plot

ret,xedge,yedge,bin = stats.binned_statistic_2d(w1_w2,bp_dd,cat["MGH"],'median',30)

#read out metallicity values using yumi's method
ind, = np.where((w1_w2>=xedge.min())&(w1_w2<=xedge.max())&(bp_dd>=yedge.min())&(bp_dd<=yedge.max()))
ix = np.searchsorted(xedge,w1_w2[ind])
ix = ix-1
iy = np.searchsorted(yedge,bp_dd[ind])
iy = iy-1
new_metal = ret[ix,iy]

#find medium, mean metallicity
mean = np.nanmean(new_metal)
median = np.nanmedian(new_metal)
print("median: ",median)
print("mean: ",mean)
    
#save color plot
fig=plt.figure(2)
ax = plt.imshow(ret.T,extent=[xedge.min(),xedge.max(),yedge.min(),yedge.max()],origin='lower',cmap='jet')
#plt.scatter(bp_rp[ind],bp_dd[ind],c=new_metal,s=2)

cbar = plt.colorbar(shrink=0.65)
cbar.set_label('[Mg/H]', rotation=270, labelpad=+16)
plt.ylabel('BP-DDO51')
plt.xlabel('W1-W2')
#plt.title('Calibration Metallicities')
plt.show()
fig.savefig('APOGEE_WISE_GAIA_2D_hist_mgh_mean.png')
#print("number of matches: ",len(ind),"out of ",len(objs_new["U"]),"stars in ",filename[19:27])
```
